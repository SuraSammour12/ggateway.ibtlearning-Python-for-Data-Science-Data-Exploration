Create a variable salary and assign it a value. Calculate a 10% bonus of the salary and print the final amount after adding the bonus to the original salary.

```python
salary = 12000
bonus_rate = 0.10
bonus_amount = salary * bonus_rate
total_amount = salary + bonus_amount
print(f'The original salary is: {salary:.2f}')
print(f'The bonus amount is: {bonus_amount:.2f}')
print(f'The final amount after bonus is: {total_amount:.2f}')
```

Write a function named calculate_area that takes the radius of a circle as an argument and returns the area of the circle. Use this function to calculate and print the area of a circle with a radius of 5.

```python
import math
def calculate_area(radius):
  if radius < 0:
    return 'Radius cannot be negative'
  area = math.pi * (radius**2)
  return area

radius = 5
area_of_circle = calculate_area(radius)
print(f'The area of the circle with a radius of {radius} is: {area_of_circle}')
```
